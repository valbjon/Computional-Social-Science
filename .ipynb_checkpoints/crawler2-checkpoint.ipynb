```python
import pandas as pd
import os, sys
from bs4 import BeautifulSoup 
from urllib.request import urlopen as uReq
import requests
import wikipedia as wiki
import timeit
```

```python
df = pd.read_csv('persons.csv')
```

```python
df
```

```python
n=df.shape[0]
n
```

```python
def crawl_records(n1, n2, tempDf):
    for i in range(n1,n2):
    
        # Get article's id
        dbpediaUrl = df.loc[i, 'uri']
        dbpediaPage= requests.get(dbpediaUrl)
        soup = BeautifulSoup(dbpediaPage.content, "html.parser")
        try:
            wikiId = soup.find(property="dbo:wikiPageID").get_text().strip()
        except AttributeError:
            wikiId = " "

        # Get article's creation date
        title = dbpediaUrl.split("/")[-1]
        historyUrl = "https://en.wikipedia.org/w/index.php?title="+title+"&dir=prev&action=history"
        historyPage = requests.get(historyUrl)
        if historyPage.status_code != 200:
            historyPage = requests.get("https://en.wikipedia.org/w/index.php?title=" + title+ "&action=history")
        soup = BeautifulSoup(historyPage.content, "html.parser")
        try:
            creationDate = soup.findAll("a", class_="mw-changeslist-date")[-1].get_text().strip().split(" ")[3]
        except Exception:
            continue

        dic = {"Id": wikiId, "CreationDate": creationDate}
        if tempDf.empty:
            tempDf = pd.DataFrame(data=dic, index=[0])
        else:
            df_temp = pd.DataFrame(data=dic, index=[0])
            tempDf = pd.concat([tempDf, df_temp])
    return tempDf
```

```python
tempDf = pd.DataFrame()
```

```python
startTime = timeit.default_timer()
tempDf = crawl_records(700000, 700100, tempDf)
elapsedTime = timeit.default_timer() - startTime
print("Crawl time ", elapsedTime)
```
