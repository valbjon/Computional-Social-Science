```python
import pandas as pd
import os, sys
from bs4 import BeautifulSoup 
from urllib.request import urlopen as uReq
import requests
import wikipedia as wiki
import timeit
```

# This crawler is adding the wikiId and creation date of the article

- It is a one-shot crawler

- It crawls about 90% of the data

```python
df = pd.read_csv('persons.csv')
```

```python
def insert_entry(tempDf, wikiId, creationDate, uri):
    """ Insert an entry to the data frame.
    
        Args:
            df: the data frame where the entry is inserted
            wikiId: the id of the inserted entry
            creationDaten: the date when the inserted entry was created
            uri: dbpediaUrl
            
        Returns:
            a data frame
    """
    dic = {"Id": wikiId, "CreationDate": creationDate, "uri": uri}
    if tempDf.empty:
        tempDf = pd.DataFrame(data=dic, index=[0])
    else:
        temp = pd.DataFrame(data=dic, index=[0])
        tempDf = pd.concat([tempDf, temp])
    return tempDf
```

```python
def crawl_records(n1, n2, tempDf):
    for i in range(n1,n2):
        print("Entered page: ", i)
        # Get article's id
        dbpediaUrl = df.loc[i, 'uri']
        try:
            dbpediaPage= requests.get(dbpediaUrl)
        except Exception:
            tempDf = insert_entry(tempDf, " ", " ", dbpediaUrl)
            continue
        soup = BeautifulSoup(dbpediaPage.content, "html.parser")
        try:
            wikiId = soup.find(property="dbo:wikiPageID").get_text().strip()
        except AttributeError:
            wikiId = " "

        # Get article's creation date
        title = dbpediaUrl.split("/")[-1]
        historyUrl = "https://en.wikipedia.org/w/index.php?title="+title+"&dir=prev&action=history"
        try:
            historyPage = requests.get(historyUrl)
        except Exception:
            tempDf = insert_entry(tempDf, " ", " ", dbpediaUrl)
            continue
        if historyPage.status_code != 200:
            historyPage = requests.get("https://en.wikipedia.org/w/index.php?title=" + title+ "&action=history")
        soup = BeautifulSoup(historyPage.content, "html.parser")
        try:
            creationDate = soup.findAll("a", class_="mw-changeslist-date")[-1].get_text().strip().split(" ", maxsplit=2)[2]
        except Exception:
            tempDf = insert_entry(tempDf, " ", " ", dbpediaUrl)
            continue
            
        tempDf = insert_entry(tempDf, wikiId, creationDate, dbpediaUrl)

    return tempDf
```

```python
tempDf = pd.DataFrame()
```

```python
startTime = timeit.default_timer()
tempDf = crawl_records(800000, 800100, tempDf)
elapsedTime = timeit.default_timer() - startTime
print("Crawl time ", elapsedTime)
```
